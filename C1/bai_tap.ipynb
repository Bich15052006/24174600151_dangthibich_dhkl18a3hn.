```python
#bài 1
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def tinh_chu_vi(self):
        return (self.dai + self.rong) * 2

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def in_thong_tin(self):
        print(f"Dài: {self.dai}, Rộng: {self.rong}")
        print(f"Chu vi: {self.tinh_chu_vi()}, Diện tích: {self.tinh_dien_tich()}")

#khởi tạo đối tượng
hcn = HinhChuNhat(5, 3)
hcn.in_thong_tin()
```

```python
# bài 2
class TS:
    def __init__(self, hoTen, toan, ly, hoa):
        self.hoTen = hoTen
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def TinhTong(self):
        return self.toan + self.ly + self.hoa

    def InTS(self):
        print("TS: {} - Tong diem: {}".format(self.hoTen, self.TinhTong()))

# khởi tạo
ds = [
    TS("Nguyen Van A", 8, 7, 6),
    TS("Le Thi B", 5, 6, 5),
    TS("Tran Van C", 9, 8, 9)
]

print("Danh sach thi sinh trung tuyen (>= 20):")
ds_trung_tuyen = sorted([ts for ts in ds if ts.TinhTong() >= 20], key=lambda x: x.TinhTong(), reverse=True)
for ts in ds_trung_tuyen:
    ts.InTS()
```

```python
# bài 4
class NganXep:
    # Thuộc tính của lớp
    SoLuongNganXep = 0

    # Hàm khởi tạo đối tượng
    def __init__(self, sucChua):
        self.sucChua = sucChua      # Sức chứa tối đa (capacity)
        self.duLieu = []            # Mảng float được cấp phát động (list trong Python)
        NganXep.SoLuongNganXep += 1
        print("Mot ngan xep moi da duoc tao voi suc chua: {}".format(sucChua))

    # Kiểm tra ngăn xếp trống
    def IsEmpty(self):
        return len(self.duLieu) == 0

    # Kiểm tra ngăn xếp đầy
    def IsFull(self):
        return len(self.duLieu) == self.sucChua

    # Thêm một phần tử vào ngăn xếp (push)
    def Push(self, giaTri):
        if not self.IsFull():
            self.duLieu.append(float(giaTri))
            print("Da dua {} vao ngan xep".format(giaTri))
        else:
            print("Ngan xep da day! Khong the dua {} vao".format(giaTri))

    # Lấy một phần tử ra từ đỉnh ngăn xếp (pop)
    def Pop(self):
        if not self.IsEmpty():
            giaTri = self.duLieu.pop()
            print("Da lay {} ra tu ngan xep".format(giaTri))
            return giaTri
        else:
            print("Ngan xep dang trong! Khong co gi de lay")
            return None

# khởi tạo
myStack = NganXep(3)
myStack.Push(15.5)
myStack.Push(20.0)
myStack.Push(25.5)
myStack.Push(30.0)
myStack.Pop()
# Kiểm tra trạng thái
if myStack.IsEmpty():
    print("Ngan xep dang trong")
else:
    print("Ngan xep van con phan tu")
```

```python
# bài 6
class NganXep:
    # Thuộc tính của lớp
    SoLuongNganXep = 0

    # Hàm khởi tạo đối tượng
    def __init__(self, sucChua):
        self.sucChua = sucChua      # Sức chứa tối đa
        self.duLieu = []            # Danh sách chứa các phần tử float
        NganXep.SoLuongNganXep += 1
        print("Mot ngan xep moi da duoc tao!")

    def IsFull(self):
        return len(self.duLieu) == self.sucChua

    def Push(self, giaTri):
        if not self.IsFull():
            self.duLieu.append(float(giaTri))
            print("Da đưa {} vao ngan xep".format(giaTri))
        else:
            print("Ngan xep da day!")

    def Print(self):
        # Kiểm tra xem ngăn xếp có dữ liệu hay không
        if len(self.duLieu) == 0:
            print("Ngan xep dang trong, khong co gi de in.")
        else:
            # In toàn bộ nội dung ngăn xếp ra màn hình
            print("Noi dung cua ngan xep la: {}".format(self.duLieu))

# khởi tạo
st = NganXep(5)
st.Push(1.1)
st.Push(2.2)
st.Push(3.3)
st.Print()
```

```python
# bài 7
class Date:
    DateCount = 0

    def __init__(self, ngay=1, thang=1, nam=2000):
        self.ngay = ngay
        self.thang = thang
        self.nam = nam
        Date.DateCount += 1
        print("Mot doi tuong Date da duoc tao!")

    def Display(self):
        print("Ngay hien tai: {:02d}/{:02d}/{:4d}".format(self.ngay, self.thang, self.nam))

    def KiemTraNamNhuan(self):
        if (self.nam % 4 == 0 and self.nam % 100 != 0) or (self.nam % 400 == 0):
            return True
        return False

    def Next(self):
        ngay_trong_thang = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
        if self.KiemTraNamNhuan():
            ngay_trong_thang[2] = 29

        self.ngay += 1 # Tăng ngày lên 1

        # Kiểm tra nếu vượt quá số ngày của tháng hiện tại
        if self.ngay > ngay_trong_thang[self.thang]:
            self.ngay = 1
            self.thang += 1
            
            # Kiểm tra nếu vượt quá tháng 12
            if self.thang > 12:
                self.thang = 1
                self.nam += 1
        
        print("Da chuyen sang ngay tiep theo!")

# khởi tạo
homNay = Date(31, 12, 2025)
homNay.Display()
homNay.Next()
homNay.Display()
```

```python
# bài 8
class Date:
    def __init__(self, ngay=1, thang=1, nam=2000):
        self.ngay = ngay
        self.thang = thang
        self.nam = nam

    def Display(self):
        # Trả về chuỗi định dạng thay vì in trực tiếp để lớp Employee dễ sử dụng
        return "{:02d}/{:02d}/{:4d}".format(self.ngay, self.thang, self.nam)

class Employee:
    # Thuộc tính của lớp
    EmployeeCount = 0

    # Hàm khởi tạo đối tượng
    def __init__(self, hoTen, ngaySinh, ngayVao):
        self.hoTen = hoTen
        self.ngaySinh = ngaySinh  # Đây là một đối tượng thuộc lớp Date
        self.ngayVao = ngayVao    # Đây cũng là một đối tượng thuộc lớp Date
        Employee.EmployeeCount += 1
        print("Mot nhan vien moi da duoc tao!")

    # Phương thức in thông tin nhân viên
    def InThongTin(self):
        print("-" * 30)
        print("Ho va ten: {}".format(self.hoTen))
        # Gọi phương thức Display() của đối tượng Date
        print("Ngay sinh: {}".format(self.ngaySinh.Display()))
        print("Ngay vao cong ty: {}".format(self.ngayVao.Display()))
        print("-" * 30)

# khởi tạo đối tượng
ngay_sinh_nv1 = Date(15, 5, 1995)
ngay_vao_nv1 = Date(1, 1, 2024)
nhanVien1 = Employee("Nguyen Van Hung", ngay_sinh_nv1, ngay_vao_nv1)
nhanVien1.InThongTin()
# Thử tạo thêm nhân viên khác
nhanVien2 = Employee("Le Thi Mai", Date(20, 10, 1998), Date(5, 6, 2025))
nhanVien2.InThongTin()
print("Tong so nhan vien hien co: {}".format(Employee.EmployeeCount))
```

```python
# bài 9
class DaGiac:
    SoLuongDaGiac = 0

    def __init__(self, tenGoi, dsCanh):
        self.tenGoi = tenGoi
        self.dsCanh = dsCanh  # Danh sách độ dài các cạnh
        DaGiac.SoLuongDaGiac += 1

    def TinhChuVi(self):
        return sum(self.dsCanh)

    def InThongTin(self):
        print("Hinh: {} - Chu vi: {}".format(self.tenGoi, self.TinhChuVi()))

# --- Kế thừa: Hình Bình Hành từ Đa Giác ---
class HinhBinhHanh(DaGiac):
    def __init__(self, canhA, canhB, chieuCao):
        # Gọi hàm khởi tạo của lớp cha (DaGiac)
        # Hình bình hành có 4 cạnh: a, b, a, b
        super().__init__("Hinh Binh Hanh", [canhA, canhB, canhA, canhB])
        self.canhA = canhA
        self.chieuCao = chieuCao

    def TinhDienTich(self):
        return self.canhA * self.chieuCao

    def InThongTin(self):
        super().InThongTin()
        print("Dien tich Hinh Binh Hanh: {}".format(self.TinhDienTich()))

# --- Kế thừa: Hình Chữ Nhật từ Hình Bình Hành ---
class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong):
        # Hình chữ nhật là Hình bình hành có chiều cao bằng cạnh bên (rong)
        super().__init__(dai, rong, rong)
        self.tenGoi = "Hinh Chu Nhat" # Cập nhật lại tên gọi

    def TinhDienTich(self):
        # Ghi đè phương thức tính diện tích (Dài x Rộng)
        return self.canhA * self.chieuCao

# --- Kế thừa: Hình Vuông từ Hình Chữ Nhật ---
class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        # Hình vuông là Hình chữ nhật có Dài = Rộng = cạnh
        super().__init__(canh, canh)
        self.tenGoi = "Hinh Vuong"
 
# khởi tạo đối tượng
hcn = HinhChuNhat(10, 5)
hcn.InThongTin()
print("-" * 20)
# Tạo một Hình Vuông
hv = HinhVuong(4)
hv.InThongTin()
```

```python
# bài 10
import math

class Diem:
    SoLuongDiem = 0

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        Diem.SoLuongDiem += 1
        print("Mot diem moi da duoc tao tai toa do ({}, {})".format(x, y))

class Elip(Diem):
    def __init__(self, x, y, trucLon, trucNho):
        # Gọi hàm khởi tạo của lớp cha (Diem) để xác định tâm Elip
        super().__init__(x, y)
        self.a = trucLon
        self.b = trucNho
        print("Mot hinh Elip da duoc tao voi ban truc lon: {}, ban truc nho: {}".format(trucLon, trucNho))

    def TinhDienTich(self):
        # Công thức diện tích Elip: pi * a * b
        return math.pi * self.a * self.b

    def InThongTin(self):
        print("Hinh Elip tai tam ({}, {})".format(self.x, self.y))
        print("Dien tich: {:.2f}".format(self.TinhDienTich()))

class DuongTron(Elip):
    def __init__(self, x, y, banKinh):
        # Đường tròn là Elip đặc biệt có hai bán trục bằng nhau và bằng bán kính r
        super().__init__(x, y, banKinh, banKinh)
        self.r = banKinh
        print("Mot hinh Tron da duoc tao voi ban kinh: {}".format(banKinh))

    def InThongTin(self):
        print("Hinh Tron tai tam ({}, {})".format(self.x, self.y))
        print("Ban kinh: {}".format(self.r))
        print("Dien tich: {:.2f}".format(self.TinhDienTich()))

# khởi tạo đối tượng
hinh_elip = Elip(0, 0, 10, 5)
hinh_elip.InThongTin()
print("-" * 30)
# Tạo một hình Tròn (kế thừa từ Elip)
hinh_tron = DuongTron(2, 3, 7)
hinh_tron.InThongTin()
```

```python
import math

class TamGiac:
    SoLuongTamGiac = 0

    def __init__(self, canhA, canhB, canhC):
        self.a = canhA
        self.b = canhB
        self.c = canhC
        TamGiac.SoLuongTamGiac += 1
        print("Mot hinh Tam Giac da duoc tao!")

    def TinhChuVi(self):
        # Công thức tính chu vi tam giác: a + b + c
        return self.a + self.b + self.c

    def InThongTin(self):
        print("Tam Giac có các canh: {}, {}, {}".format(self.a, self.b, self.c))
        print("Chu vi: {}".format(self.TinhChuVi()))

# --- Kế thừa: Tam Giác Cân từ Tam Giác ---
class TamGiacCan(TamGiac):
    def __init__(self, canhBen, canhDay):
        # Tam giác cân có hai cạnh bên bằng nhau
        super().__init__(canhBen, canhBen, canhDay)
        print("Day la mot Tam Giac Can")

# --- Kế thừa: Tam Giác Vuông từ Tam Giác ---
class TamGiacVuong(TamGiac):
    def __init__(self, canhGocVuong1, canhGocVuong2):
        # Tính cạnh huyền theo định lý Pytago
        canhHuyen = math.sqrt(canhGocVuong1**2 + canhGocVuong2**2)
        super().__init__(canhGocVuong1, canhGocVuong2, canhHuyen)
        print("Day la mot Tam Giac Vuong")

# --- Kế thừa: Tam Giác Đều từ Tam Giác Cân ---
class TamGiacDeu(TamGiacCan):
    def __init__(self, canh):
        # Tam giác đều là tam giác cân có cạnh bên bằng cạnh đáy
        super().__init__(canh, canh)
        print("Day la mot Tam Giac Deu")

# khởi tạo đối tượng
# 1. Thử nghiệm Tam Giác Vuông
tg_vuong = TamGiacVuong(3, 4)
tg_vuong.InThongTin()
print("-" * 30)
# 2. Thử nghiệm Tam Giác Đều
tg_deu = TamGiacDeu(6)
tg_deu.InThongTin()
print("-" * 30)
print("Tong so hinh tam giac da tao: {}".format(TamGiac.SoLuongTamGiac))
```
