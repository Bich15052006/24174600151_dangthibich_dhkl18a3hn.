```python
class Dog:

    def __init__(self, name, size, age, color):   
        self._name = name  #thuộc tính protected
        self._size = size  #thuộc tính protected
        self.__age = age   # thuộc tính private
        self.__color = color  #thuộc tính private

    @property  # đn pthuc getter cho thuộc tính private __age
    def age(self):
        print("Getter for age called")
        if not hasattr(self, "_Dog__age"):  #kiểm tra xem __age có tồn tại không
            raise ArithmeticError("Age has been deleted or not set")
        return self.__age  #trả về vị trí tuổi
    
    @age.setter
    def age(self, value):  #setter cho thuộc tính private__age
        print("Setter for age called")
        if value < 0:
            raise ValueError("Age cannot be negative")
        self.__age = value  #cập nhật giá trị tuổi private__age

    @age.deleter  #đn deleter cho thuộc tính private__age 
    def age(self):
        print("Deleter for age called")
        if not hasattr(self, "_Dog__age"):
            raise ArithmeticError("Age has already been deleted or not set")
        del self.__age  #xóa thuộc tính private__age
```

```python
class Animal:
    def __init__(self, name):
        self._name = name

    def Display(self):
        print("I'm {}".format(self._name))

class Dog(Animal):
    def __init__(self, name, size, age, color):
        super().__init__(name)
        self.size = size
        self.age = age
        self.color = color

    def Go(self, place):
        print("I'm going to {}".format(place))

obj1 = Dog("bull", "large", 2, "yellow")
obj1.Display()
obj1.Go("garden")
```

```python
class Zebra:
    def __init__(self):
        print("Zebra")

    def Display(self):
        print("I'm a zebra")

class Donkey:
    def __init__(self):
        print("Donkey")

    def Display(self):
        print("I'm a donkey")

class Zonkey(Zebra, Donkey):
    def __init__(self):
        Zebra.__init__(self)
        Donkey.__init__(self)
        print("Zonkey")

obj = Zonkey()
obj.Display()
```

```python
class LopCha:
    def __init__(self, ten_cha):
        self.ten_cha = ten_cha

    def thong_tin(self):
        print("Tôi là Lớp Cha, tên của tôi là", self.ten_cha)

class LopCon(LopCha):
    def __init__(self, ten_cha, ten_con):
        super().__init__(ten_cha)
        self.ten_con = ten_con

    def thong_tin(self):
        print("Tôi là Lớp Con, tên của tôi là", self.ten_con)
        super().thong_tin()

class LopChau(LopCon):
    def __init__(self, ten_cha, ten_con, ten_chau):
        super().__init__(ten_cha, ten_con)
        self.ten_chau = ten_chau

    def thong_tin(self):
        print("Tôi là Lớp Cháu, tên của tôi là", self.ten_chau)
        super().thong_tin()

lop_chau = LopChau("Ông A", "Bác B", "Cháu C")
lop_chau.thong_tin()
```

```python
#khai báo lớp chim
class Bird:
    def flight(self): #đn pthuc flight()
        print("Many birds can fly")
#khai báo lớp chim sẻ kế thừa từ lớp Bird
class Sparrow(Bird):
    def flight(self):  #ghi đè pthuc flight
        print("Sparrow can fly")

class Ostrich(Bird):
    def flight(slef):
        print("Ostriches cannot fly")  #pthuc flight được ghi đè (overriding), tahy đổi nội dung dòng lệnh print đề ra:"Ostriches cannot fly"

obj1 = Bird()
obj2 = Sparrow()
obj3 = Ostrich()
obj1.flight()
obj2.flight()
obj3.flight()
```

```python
from abc import ABC, abstractmethod
#tạo lớp trừu tượng 'Animal' với pthuc trừu tượng 'Say'
class Animal(ABC):
    @abstractmethod
    def Say(self):
        pass

class Dog(Animal):
    def Say(self):
        print("Whoooopp whoopp...")

class Cat(Animal):
    def Say(self):
        print("Meoo meoo")
    #đn 1 pthuc 'Jump' cho lớp 'Cat'
    def Jump(self):
        print("I'm jumping up high")
#khởi tạo đtg từ lơps 'Dog'
obj_dog = Dog()

obj_dog.Say()

obj_cat=Cat()

obj_cat.Say()
```
