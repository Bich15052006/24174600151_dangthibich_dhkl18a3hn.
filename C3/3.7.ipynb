```python
import numpy as np

```

# Bước 0: Đọc dữ liệu từ file baseball_2D.txt


```python

np_baseball = np.loadtxt("baseball_2D.txt")
print("Bước 0: Đã đọc file baseball_2D.txt")
```

# Bước 1: Kiểu dữ liệu và kích thước


```python

print("Bước 1:")
print("Kiểu dữ liệu:", np_baseball.dtype)
print("Kích thước:", np_baseball.shape)
```

# Bước 2: In dòng thứ 50 (index 49)


```python

print("\nBước 2: Dòng thứ 50:", np_baseball[49])
```

# Bước 3: Tạo np_weight từ cột 2


```python

np_weight = np_baseball[:, 1]
print("\nBước 3: np_weight:", np_weight)
```

# Bước 4: Chiều cao vận động viên thứ 124 (index 123)


```python
print("\nBước 4: Chiều cao vận động viên thứ 124:", np_baseball[123, 0])
```

# Bước 5: Chiều cao và cân nặng trung bình


```python
avg_height = np.mean(np_baseball[:, 0])
avg_weight = np.mean(np_baseball[:, 1])
print("\nBước 5:")
print("Chiều cao trung bình:", round(avg_height, 2), "inch")
print("Cân nặng trung bình:", round(avg_weight, 2), "pound")
```

# Bước 6: Nhận xét về tương quan chiều cao - cân nặng


```python
correlation = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("\nBước 6: Hệ số tương quan chiều cao - cân nặng:", round(correlation, 3))
if correlation > 0.5:
    print("Nhận xét: Có tương quan thuận mạnh giữa chiều cao và cân nặng.")
elif correlation > 0:
    print("Nhận xét: Có tương quan thuận yếu giữa chiều cao và cân nặng.")
elif correlation < -0.5:
    print("Nhận xét: Có tương quan nghịch mạnh giữa chiều cao và cân nặng.")
elif correlation < 0:
    print("Nhận xét: Có tương quan nghịch yếu giữa chiều cao và cân nặng.")
else:
    print("Nhận xét: Không có tương quan giữa chiều cao và cân nặng.")
```
