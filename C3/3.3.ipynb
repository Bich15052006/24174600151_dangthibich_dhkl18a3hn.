```python
import numpy as np

```

# Khởi tạo 2 array arr_a và arr_b

```python
arr_a = np.array([1,2,3,2,3,4,3,4,5,6])
arr_b = np.array([7,2,10,2,7,4,9,4,9,8])
```

# 1. Tạo arr_c chỉ lấy duy nhất các phần tử xuất hiện ở cả arr_a và arr_b

```python
arr_c = np.intersect1d(arr_a, arr_b)
print("arr_c (xuất hiện cả arr_a và arr_b):", arr_c)
```

# 2. Tạo arr_d chứa các phần tử chỉ xuất hiện ở arr_a


```python
arr_d = np.setdiff1d(arr_a, arr_b)
print("arr_d (chỉ xuất hiện trong arr_a):", arr_d)
```

# 3. Tạo arr_f chứa các phần tử có giá trị từ 5 đến 10 của arr_e

```python
arr_e = np.array([2, 6, 1, 9, 10, 3, 27, 8, 6, 25, 16])
arr_f = arr_e[(arr_e >= 5) & (arr_e <= 10)]
print("arr_f (giá trị từ 5 đến 10):", arr_f)
```
