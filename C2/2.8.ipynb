```python
import json

```

#Ví dụ đối tượg python

```python

doi_tuong_python = {
    "ten": "Nguyen Van C",
    "tuoi": 22,
    "lop": "11A4",
    "diem": [7.5, 8.0, 9.0]
}
```

#Chuyển thành chuỗi json

```python

json_str = json.dumps(doi_tuong_python)
```

#In ra chuỗi json

```python

print("Chuỗi JSON từ đối tượng Python:")
print(json_str)

```

#In ra tất cả các giá trị

```python

print("\nCác giá trị của đối tượng:")
for key, value in doi_tuong_python.items():
    print(f"{key}: {value}")
```
