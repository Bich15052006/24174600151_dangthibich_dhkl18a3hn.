```python
import json
from datetime import datetime

# Danh sách lưu 
giao_dich_list = []

print("Chương trình quản lý giao dịch tiền và vàng")
while True:
    loai = input("Nhập loại giao dịch (tien/vang): ").strip().lower()
    so_luong = input("Nhập số lượng: ").strip()
    gia_tri = input("Nhập giá trị: ").strip()
    
    giao_dich = {
        "loai": loai,
        "so_luong": so_luong,
        "gia_tri": gia_tri
    }
    giao_dich_list.append(giao_dich)
    
    tiep_tuc = input("Bạn có muốn tiếp tục giao dịch không? (c/k): ").strip().lower()
    if tiep_tuc != 'c':
        break

# Hỏi người dùng có muốn ghi vào file không
ghi = input("Bạn có muốn ghi các giao dịch vào file không? (1: Có, 0: Không): ").strip()
if ghi == '1':
    # Lấy thời gian hiện tại
    now = datetime.now()
    filename = now.strftime("%Y-%m-%d-%H-%M-%S") + ".json"
    
    # Ghi danh sách giao dịch vào file JSON
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(giao_dich_list, f, ensure_ascii=False, indent=4)
    
    print(f"Đã ghi các giao dịch vào file {filename}")
else:
    print("Không ghi dữ liệu vào file.")
```
