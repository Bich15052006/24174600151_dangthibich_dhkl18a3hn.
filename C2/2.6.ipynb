```python
import urllib.request
import xml.etree.ElementTree as ET
import csv

# RSS XML hợp lệ
url = "https://www.hindustantimes.com/feeds/rss/top-news.xml"

# 1. Tải RSS (KHÔNG dùng requests)
with urllib.request.urlopen(url) as response:
    data = response.read()

print("Tải RSS feed thành công")

# 2. Parse XML
root = ET.fromstring(data)

tin_tuc_list = []

for item in root.findall(".//item"):
    title = item.findtext("title", "")
    link = item.findtext("link", "")
    pubDate = item.findtext("pubDate", "")

    tin_tuc_list.append({
        "title": title,
        "link": link,
        "pubDate": pubDate
    })

print("Đã phân tích", len(tin_tuc_list), "mục tin")

# 3. Ghi CSV
with open("tin_tuc.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "link", "pubDate"])
    writer.writeheader()
    writer.writerows(tin_tuc_list)

print("Đã lưu tin tức vào tin_tuc.csv")
```
