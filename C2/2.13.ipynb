```python
import requests

# URL API lấy comment của post có id = 1
url = "https://jsonplaceholder.typicode.com/comments?postId=1"

# Gửi yêu cầu GET tới API
response = requests.get(url)

if response.status_code == 200:
    data = response.json()  # Chuyển JSON thành danh sách Python
else:
    print("Lỗi khi truy cập API:", response.status_code)
    exit()

# Giới hạn chỉ hiển thị 3 bài đầu
data_limit = data[:3]

print("Danh sách các bài post nổi bật (postId=1, 3 bài đầu):\n")

for i, bai in enumerate(data_limit, start=1):
    print(f"Bài post {i}:")
    print(f"Post ID: {bai['postId']}")
    print(f"ID: {bai['id']}")
    print(f"Name: {bai['name']}")
    print(f"Email: {bai['email']}")
    print(f"Body: {bai['body']}\n")
```
