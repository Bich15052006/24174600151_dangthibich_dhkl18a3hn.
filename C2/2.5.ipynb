```python
from xml.dom import minidom

```

#1.Nạp file xml

```python

doc = minidom.parse('sample.xml')
```

#2.Lấy tất cả các phần tử

```python

danh_sach_phan_tu = doc.getElementsByTagName('*')  # '*' lấy tất cả các phần tử
```

#3.In ra các phần tử

```python

print("Danh sách tên các phần tử:")
for phan_tu in danh_sach_phan_tu:
    print(phan_tu.tagName)
```
