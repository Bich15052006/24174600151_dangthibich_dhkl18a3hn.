```python
import json
```

#Ví dụ dữ liệu json

```python

json_data = '''
{
    "ten": "Nguyen Van A",
    "tuoi": 25,
    "diem": [8, 9, 7.5],
    "lop": "12A1"
}
'''
```

#Chuyển json thành đối tượng python

```python

obj = json.loads(json_data)

print("Kiểu dữ liệu:", type(obj))
print("Dữ liệu Python:", obj)
print("Tên:", obj['ten'])
print("Tuổi:", obj['tuoi'])
print("Điểm:", obj['diem'])
```
