```python
import json 

```

#Ví dụ đối tượng dict python

```python

doi_tuong_dict = {
    "ten": "Tran Van Bon",
    "tuoi": 23,
    "lop": "10A3",
    "diem": [8.5, 7.0, 6.0]
}
```

#Chuyển dict thành JSON, sắp xếp theo key, thụt lề 4

```python
json_str_sorted = json.dumps(doi_tuong_dict, sort_keys=True, indent=4)
```

#In ra JSON đẹp với thụt lề


```python
print("JSON sắp xếp theo khóa, thụt lề 4:")
print(json_str_sorted)
```
