```python
import json
```

#Đọc dữ liệu từ file json

```python

with open('nhanvien.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

ten_cong_ty = data['ten_cong_ty']
dia_chi = data['dia_chi']
nhan_vien = data['nhan_vien']
```

#Tổng số nhân viên

```python

tong_nv = len(nhan_vien)
```

#Thống kê theo đơn vị

```python

don_vi_dict = {}
for nv in nhan_vien:
    dv = nv['don_vi']
    if dv not in don_vi_dict:
        don_vi_dict[dv] = 0
    don_vi_dict[dv] += 1
```

#In kết quả

```python

print(f"Tên công ty: {ten_cong_ty}")
print(f"Địa chỉ: {dia_chi}")
print("----Thống kê nhân viên theo đơn vị----")

for i, (dv, so_nv) in enumerate(don_vi_dict.items(), start=1):
    ty_le = (so_nv / tong_nv) * 100
    print(f"{i}./Tên đơn vị: {dv}")
    print(f"- Số nhân viên: {so_nv}")
    print(f"- Tỷ lệ: {ty_le:.2f}%\n")
```
