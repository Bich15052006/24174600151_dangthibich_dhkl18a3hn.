```python
nv_location=pd.DataFrame(
{
'Nhân viên':['Bình', 'Minh','Lan'],
'Nơi sinh':['Hà Nội','Hồ Chí Minh','Nam định']
}
)
nv_age=pd.DataFrame(
{
'Nhân viên':['Bình','Vân','Lan'],
'Age':[22,23,25]
}
)
```

```python
nv_location
```

```python
nv_age
```

```python
df_inner=pd.merge(nv_location, nv_age, on='Nhân viên', how='inner')
print('INNER:'); display(df_inner)
```

```python
df_outer=pd.merge(nv_location, nv_age, on='Nhân viên', how='outer')
print('OUTER:'); display(df_outer)
```

```python
df_left=pd.merge(nv_location, nv_age, on='Nhân viên', how='left')
print('LEFT:'); display(df_left)
```

```python
vang_rank=pd.DataFrame({'Nhân viên': ['Bình','Lan','Hà'],'rank':[1,2,3]})
hs_rank=pd.DataFrame({'Nhân viên': ['Bình','Lan','Hà'],'rank':[4,5,6]})
```

```python
vang_rank
```

```python
hs_rank
```

```python
default_suffix=pd.merge(vang_rank, hs_rank, on='Nhân viên')
default_suffix
```

```python
custom_suffix = pd.merge(vang_rank, hs_rank, on='Nhân viên', suffixes=('_Ngày vắng','_Hiệu suất'))
print('Mặc định (x, y):'); display(default_suffix)
print('Tùy chỉnh suffixes:'); display(custom_suffix)
```

```python
import pandas as pd
import seaborn as sns
titanic_data = sns.load_dataset("titanic")
titanic_data.head()
```

```python
titanic_data.groupby("sex")["age"].mean()
```

```python
gb_age=titanic_data.groupby(["sex", "class"])["age"].mean()
gb_age_unstack=gb_age.unstack()
gb_age_unstack
```

```python
pivot_age = pd.pivot_table(titanic_data,values="age",index="sex",columns="class",aggfunc="mean",observed=True)
pivot_age
```

```python
gb_age_unstack
```

```python
titanic_data.isna().sum().sort_values(ascending=False)
```
