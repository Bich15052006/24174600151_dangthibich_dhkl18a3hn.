```python
import numpy as np
import time
arr = np.arange(100000000, dtype=np.int64)
start_time = time.time()
total = np.int64(0)
for num in arr:
    total += num
end_time = time.time()
elapsed_time_loop = end_time - start_time
print("Tổng các phần tử mảng arr khi sử dụng vòng lặp:", total)
print("Thời gian (giây) khi sử dụng vòng lặp:", elapsed_time_loop)
start_time = time.time()
total = np.sum(arr)
end_time = time.time()
elapsed_time_npsum = end_time - start_time
print("Tổng các phần tử mảng arr khi sử dụng hàm np.sum():", total)
print("Thời gian (giây) khi sử dụng np.sum():", elapsed_time_npsum)
num_times_faster = elapsed_time_loop / elapsed_time_npsum
print("Sử dụng vòng lặp chậm hơn: ~",int(num_times_faster),"lần")
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
# Căn bậc hai của các phần tử trong mảng
kq_can_hai = np.sqrt(arr) 
# Lũy thừa 2 của các phần tử trong mảng
kq_luy_thua_2 = np.power(arr, 2)
# Logarit tự nhiên (cơ số e) của các phần tử trong mảng
kq_logarit = np.log(arr)
# Giá trị tuyệt đối của các phần tử trong mảng
kq_tri_tuyet_doi = np.abs(arr)
# Hàm logic NOT (đảo ngược giá trị) của các phần tử trong mảng
kq_not_logic = np.logical_not(arr)
#Hàm sin của các phần tử trong mảng
kq_sin=np.sin(arr)
print("Căn bậc hai của từng phần tử trong mảng là: ", kq_can_hai) 
print("Lũy thừa bậc hai của từng phần tử trong mảng là: ", kq_luy_thua_2)
print("Logarit của từng phần tử trong mảng là: ", kq_logarit) 
print("Trị tuyệt đối của từng phần tử trong mảng là: ", kq_tri_tuyet_doi) 
print("Kết quả đảo ngược giá trị của các phần tử trong mảng là:", kq_not_logic)
print("Kết quả hàm sin của các phần tử trong mảng là:", kq_sin)
```

```python
import pandas as pd
indA=pd.Index([0, 2, 4, 6, 8])
indB=pd.Index([1, 3, 7, 8, 9])
print('Index A: ', indA)
print('Index B: ', indB)
print('Hợp: ', indA.union(indB)) #hợp (union)
print('Hiệu: ', indA.difference(indB)) #hiệu
print('Giao: ', indA.intersection(indB)) #Giao
print('Hiệu đối xứng: ',indA.symmetric_difference(indB))#hiệu đối xứng
print('A <= B and A!=B: ', indA.isin(indB))
```
