```python
import pandas as pd
```

# Input

```python
ser = pd.Series([1, 3, 6, 10, 15, 21, 27, 35])
```

# Hiệu số giữa các số liên tiếp (first difference)


```python
diff1 = ser.diff()
print("Hiệu số bậc 1:")
print(diff1)
```

# Hiệu số của hiệu số (second difference)

```python
diff2 = diff1.diff()
print("\nHiệu số bậc 2:")
print(diff2)
```
