```python
import pandas as pd
import numpy as np
```

# Tạo Series với phân phối chuẩn


```python
np.random.seed(42)  # để kết quả có thể lặp lại
ser = pd.Series(np.random.normal(10, 5, 25))
```

# 1. Giá trị tối thiểu

```python
min_val = ser.min()
print("Minimum:", min_val)
```

# 2. 25th percentile



```python
q25 = ser.quantile(0.25)
print("25th percentile:", q25)
```

# 3. Trung vị



```python
median_val = ser.median()
print("Median:", median_val)
```

# 4. 75th percentile


```python
q75 = ser.quantile(0.75)
print("75th percentile:", q75)
```

# 5. Giá trị tối đa


```python
max_val = ser.max()
print("Maximum:", max_val)
```
