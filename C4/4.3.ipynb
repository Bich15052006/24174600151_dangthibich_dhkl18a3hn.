```python
import pandas as pd
import numpy as np
```

# Input



```python
ser1 = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))
```

# Nối thành DataFrame


```python
df = pd.concat([ser1, ser2], axis=1)
df.columns = ['letter', 'number']  # Đổi tên cột

print(df.head())
```
